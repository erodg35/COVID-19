```
# Creates lists for sorting and functions

#TODO: create "stop word link" list to exclude menu links

topTerms = []
allLinks = []
news = []
chinaLinks =[]
nKLinks = []
cyberLinks = []
indiaLinks = []
covidLinks = []

def getNews(url):
  print("")
  print(url)
  print("----------")
  
  try:
    r = requests.get(url,verify=False,timeout=1).text
    status = requests.get(url)
    
    if status.status_code == 200:
      soup = bs(r)
      
      exclude_items = [" ","#","/",None]

      for link in soup.find_all('a'):
        if link.text != "" and link.text != None and link.text != " " and link.text !="#":
          news.append(link.text)
          allLinks.append([link.text.strip(),link.get('href'),link,url])
          if "china" in link.text or "China" in link.text or "Chinese" in link.text:
            chinaLinks.append([link.text.strip(),link.get('href'),link,url])
            print("China observed")
          if "cyber" in link.text or "Cyber" in link.text or "hack" in link.text:
            cyberLinks.append([link.text.strip(),link.get('href'),link,url])
            print("Cyber observed")
          if "north korea" in link.text or "North Korea" in link.text or "DPRK" in link.text:
            nKLinks.append([link.text.strip(),link.get('href'),link,url])
            print("nK observed")
          if "india" in link.text or "India" in link.text or "Indian" in link.text:
            indiaLinks.append([link.text.strip(),link.get('href'),link,url])
            print("India observed")
          if "covid" in link.text or "COVID" in link.text or "coronavirus" in link.text or "covid-19" in link.text or "COVID-19" in link.text or "cov2" in link.text or "cov2" in link.text:
            covidLinks.append([link.text.strip(),link.get('href'),link,url])
            print("COVID-19 observed")
        else:
          pass
  except: pass
```

```
#This block runs the function, will take a while to execute
for url in crawl.siteURL:
  getNews(url)
```

```
#Organizes links and text
dfLinks = pd.DataFrame(allLinks)
dfChina = pd.DataFrame(chinaLinks)
dfCyber = pd.DataFrame(cyberLinks)
dfnK = pd.DataFrame(nKLinks)
dfIndia = pd.DataFrame(indiaLinks)
dfCovid = pd.DataFrame(covidLinks)

#dfLinks.to_excel(str(today) + '-links.xlsx')
```

```
#Writes and exports to XLSX.
#Document will save to files in left menu
writer = pd.ExcelWriter(str(today) + '-news.xlsx', engine='xlsxwriter')

dfLinks.to_excel(writer,sheet_name='links')
dfChina.to_excel(writer,sheet_name='china')
dfCyber.to_excel(writer,sheet_name='cyber')
dfnK.to_excel(writer,sheet_name='dprk')
dfIndia.to_excel(writer,sheet_name='india')
dfCovid.to_excel(writer,sheet_name='covid')

writer.save()
```
